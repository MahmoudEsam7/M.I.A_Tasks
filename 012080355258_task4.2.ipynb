```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python

class LinearRegression:

    def __init__(self, learning_rate=0.000001,n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None
    def fit(self, x,y): 
        n_samples = x.shape[0]
        self.weights = np.zeros(1)
        self.bias = 0
        for _ in range (self.n_iters):
            y_pred = np.multiply(x,self.weights) + self.bias
        
            dw =(1/n_samples)* np.dot(x.T,(y_pred - y))
            dp =(1/n_samples)* np.sum(y_pred - y)

            self.weights = self.weights - self.learning_rate*dw
            self.bias = self.bias - self.learning_rate*dp
    def prediction(self,x):
        y_pred = np.multiply(x, self.weights) + self.bias
        return y_pred
```

```python

dataset = pd.read_csv('shrink_ray_dataset.csv')
x = dataset['Power']
y = dataset['Shrinkage']
```

```python

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2)

model = LinearRegression()
model.fit(x_train,y_train)
y_predit_line = model.prediction(x)
```

```python
fig = plt.figure(figsize=(8,6))
plt.scatter(x,y,color="b",marker="o",s=30)
plt.title('DataSet')
plt.xlabel('Power(MW)')
plt.ylabel('Shrinkage(%)')
plt.show 
```
